```python
nlp = spacy.load('el_core_news_sm')
```

```python
print(sorted(nlp.Defaults.stop_words)[0:20])
```

```python
def create_wordcloud(text):
  wordcloud = WordCloud(
      stopwords = nlp.Defaults.stop_words,
      width = 2000,
      height = 1000,
      background_color = 'black'
  ).generate(text)
  return wordcloud
```

```python
wcloud = create_wordcloud(july_text)
list(wcloud.words_.items())[:5]
```

```python
def create_wordcloud_img(text):
  wordcloud = create_wordcloud(text)
  fig = plt.figure(
      figsize = (40, 30),
      facecolor = 'k',
      edgecolor = 'k')
  return plt.imshow(wordcloud)
```

```python
create_wordcloud_img(july_text)
```

```python
nlp.Defaults.stop_words.add("πυρκαγιές")
print(nlp.vocab['πυρκαγιές'].is_stop)
```

```python
nlp.Defaults.stop_words.add("πυρκαγιά")
nlp.Defaults.stop_words.add("φωτιές")
nlp.Defaults.stop_words.add("φωτιά")
```

```python
create_wordcloud_img(july_text)
```

```python
july_text_lower = july_text.lower()
create_wordcloud_img(july_text_lower)
```

```python
july_doc = nlp(july_text_lower)
for token in july_doc[:10]:
    print(f"{token.text} -> {token.lemma_} ({token.pos_})")
```

```python
tokenized_july_doc = ' '.join(token.lemma_ for token in july_doc)
```
